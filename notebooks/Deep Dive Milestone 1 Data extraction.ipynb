```python
issues = {}

# Check if there is numerical issue in house price
if "price_paid" in df:
    issues["price_negative"] = int((df["price_paid"] < 0).sum())
    issues["price_zero"]     = int((df["price_paid"] == 0).sum())
    print("\nprice_paid summary:")
    print(df["price_paid"].describe(percentiles=[.01,.05,.5,.95,.99]))

# Check if there is issue with the date
if "date" in df:
    min_ok, max_ok = pd.Timestamp("1995-01-01"), pd.Timestamp.today()
    issues["date_too_early"] = int((df["date"] < min_ok).sum())
    issues["date_in_future"] = int((df["date"] > max_ok).sum())
    print("\nDate range:", df["date"].min(), "->", df["date"].max())
```

```python
# Check if there is issue with these categorical variables
allowed = {
    "property_type": {"D","S","T","F","O"},
    "new_build": {"Y","N"},
    "tenure": {"F","L"},
    "transaction_category": {"A","B"},
}
for col, ok in allowed.items():
    if col in df:
        bad = ~df[col].isin(list(ok)) & df[col].notna()
        issues[f"{col}_invalid"] = int(bad.sum())
```

```python
# Check if there is issue in postcode (British postcode format)
if "postcode" in df:
    pc_pattern = re.compile(r"^[A-Z]{1,2}\d[A-Z\d]?\s*\d[A-Z]{2}$")
    bad_pc = (df["postcode"].dropna()
                            .str.upper()
                            .str.fullmatch(pc_pattern)
                            .fillna(True)
                            .eq(False)
                            .sum())
    issues["postcode_invalid"] = int(bad_pc)

print("\n issue counts:")
```

# Merge EPC structural characteristics (avg area, avg rooms),
# sourced from the official UK EPC Register (Ministry of Housing).
# Aggregated by postcode and joined to the transaction dataset.


```python
# 1. Set file path
epc_path = "/content/drive/MyDrive/Deep Dive 12/ppd/london_epc_filtered.csv"

# 2. Load and clean EPC data
print("Loading area features...")
df_epc = pd.read_csv(epc_path, usecols=['POSTCODE', 'TOTAL_FLOOR_AREA', 'NUMBER_HABITABLE_ROOMS'])
df_epc['join_key'] = df_epc['POSTCODE'].str.replace(' ', '').str.upper()

# 3. Aggregate and deduplicate by postcode
epc_grouped = df_epc.groupby('join_key').agg({
    'TOTAL_FLOOR_AREA': 'mean',
    'NUMBER_HABITABLE_ROOMS': 'mean'
}).reset_index()
epc_grouped.columns = ['join_key', 'avg_area', 'avg_rooms']

# 4. Prepare the main dataset (ensure your main table is named 'dataset')
if 'dataset' in locals():
    dataset['join_key'] = dataset['postcode'].str.replace(' ', '').str.upper()

    # 5. Merge EPC features into the main dataset
    dataset = pd.merge(dataset, epc_grouped, on='join_key', how='left')

    # 6. Fill missing values using the mean of the merged dataset
    dataset['avg_area'] = dataset['avg_area'].fillna(dataset['avg_area'].mean())
    dataset['avg_rooms'] = dataset['avg_rooms'].fillna(dataset['avg_rooms'].mean())

    print("Feature engineering completed!")
    print("Current dataset shape:", dataset.shape)
    print("Preview of area features:")
    print(dataset[['postcode', 'avg_area', 'avg_rooms']].head())

else:
    print("Error: variable 'dataset' does not exist. Please load your PKL file first!")
```

#We merge property-level transactions with macroeconomic indicators — including population, earnings, births, inactivity rates, and affordability ratios — to enrich the dataset with borough-level fundamentals from London Datastore

```python
import os

# Ensure drive_path is defined and points to your macro data directory
drive_path = "/content/drive/MyDrive/Deep Dive 12/ppd/"

# Names of the 33 London boroughs (used for cleaning and filtering)
LONDON_BOROUGH_NAMES = [
    'City of London', 'Westminster', 'Kensington and Chelsea', 'Hammersmith and Fulham',
    'Wandsworth', 'Lambeth', 'Southwark', 'Tower Hamlets', 'Hackney', 'Islington',
    'Camden', 'Brent', 'Ealing', 'Hounslow', 'Richmond upon Thames', 'Kingston upon Thames',
    'Merton', 'Sutton', 'Croydon', 'Bromley', 'Lewisham', 'Greenwich', 'Bexley',
    'Havering', 'Barking and Dagenham', 'Barnet', 'Redbridge', 'Newham', 'Enfield',
    'Haringey', 'Harrow', 'Hillingdon', 'Waltham Forest'
]

def clean_and_melt_csv(file_path: str, skip_rows: int, measure_name: str, lad_prefix_pattern: str) -> pd.DataFrame:
    """
    Reads a CSV file, skips descriptive header rows, cleans Local Authority names,
    and converts the data into long format.

    Args:
        file_path (str): Path to the CSV file.
        skip_rows (int): Number of rows to skip before the header.
        measure_name (str): Name of the metric to store in the output column.
        lad_prefix_pattern (str): Regex pattern to match and remove LAD prefixes.

    Returns:
        pd.DataFrame: Long-format DataFrame (Area, Year, Measure).
    """
    # 1. Load CSV, skipping descriptive header rows
    df = pd.read_csv(file_path, skiprows=skip_rows, header=0, low_memory=False)

    # 2. Rename the first column to Area
    df.rename(columns={df.columns[0]: 'Area'}, inplace=True)

    # 3. Filter rows containing LAD prefixes
    filtered_df = df[df['Area'].astype(str).str.contains(lad_prefix_pattern, case=False, na=False)].copy()

    # Remove prefixes and clean borough names
    filtered_df['Area'] = (
        filtered_df['Area']
        .astype(str)
        .str.replace(lad_prefix_pattern, '', regex=True)
        .str.strip()
    )

    # 4. Identify year columns and melt to long format
    year_cols = [col for col in filtered_df.columns if str(col).isdigit() and 2015 <= int(col) <= 2024]

    melted_df = filtered_df.melt(
        id_vars=['Area'],
        value_vars=year_cols,
        var_name='Year',
        value_name=measure_name
    )

    # 5. Clean value types
    melted_df[measure_name] = pd.to_numeric(
        melted_df[measure_name].astype(str).str.replace(r'[#,-,[\[x\]]', '', regex=True),
        errors='coerce'
    )
    melted_df['Year'] = pd.to_numeric(melted_df['Year'], errors='coerce').astype('Int64')

    melted_df = melted_df.dropna(subset=[measure_name]).reset_index(drop=True)

    # Aggregate duplicates if necessary
    if not melted_df[['Area', 'Year']].duplicated().any():
        return melted_df
    else:
        return melted_df.groupby(['Area', 'Year'], as_index=False)[measure_name].mean()


def build_earn_long(path):
    """Build long-format earnings data."""
    xls = pd.ExcelFile(path)
    df = pd.read_excel(xls, sheet_name="FT workers annual Median")
    df = df.dropna(subset=[df.columns[0]])

    df = df.rename(columns={df.columns[0]: "Code", df.columns[1]: "Area"})

    # Standardize area names and filter to London boroughs
    df['Area'] = df['Area'].astype(str).str.replace('City OF LONDON', 'City of London').str.strip()
    df = df[df['Area'].isin(LONDON_BOROUGH_NAMES)].copy()

    earn_long = df.melt(
        id_vars=["Code", "Area"],
        var_name="Year",
        value_name="Median_Annual_Earnings"
    )

    earn_long["Year"] = earn_long["Year"].astype(int)
    earn_long["Median_Annual_Earnings"] = pd.to_numeric(
        earn_long["Median_Annual_Earnings"].astype(str).str.replace(r'[#,-,[\[x\]]', '', regex=True),
        errors="coerce"
    )
    earn_long = earn_long.dropna(subset=["Median_Annual_Earnings"])

    if not earn_long[['Area', 'Year']].duplicated().any():
        return earn_long
    else:
        return earn_long.groupby(['Area', 'Year'], as_index=False)["Median_Annual_Earnings"].mean()


def build_inact_long(path):
    """Build long-format economic inactivity data."""
    raw = pd.read_excel(path, sheet_name="All persons", header=None)

    header_rows = raw.iloc[0:2]

    new_columns = []
    current_period = None

    for i in range(len(header_rows.columns)):
        if i < 2:
            new_columns.append(header_rows.iloc[1, i].lower())
        else:
            if pd.notna(header_rows.iloc[0, i]):
                current_period = header_rows.iloc[0, i]
            column_type = header_rows.iloc[1, i]
            new_columns.append((current_period, column_type))

    data = raw.iloc[3:].copy()
    data.columns = new_columns

    data = data.dropna(subset=['code'])
    data = data.rename(columns={'area': 'Area'})
    data['Area'] = data['Area'].astype(str).str.replace('City OF LONDON', 'City of London').str.strip()
    data = data[data['Area'].isin(LONDON_BOROUGH_NAMES)].copy()

    percent_cols = [col for col in data.columns if isinstance(col, tuple) and col[1] == "percent"]

    records = []

    for (period, type_name) in percent_cols:
        m = re.search(r"\d{4}", str(period))
        if not m:
            continue
        year = int(m.group())

        tmp = data[["code", "Area", (period, type_name)]].copy()
        tmp.columns = ["code", "Area", "Inactivity_Rate_Percent"]
        tmp["Year"] = year
        records.append(tmp)

    inact_long = pd.concat(records, ignore_index=True)

    inact_long["Inactivity_Rate_Percent"] = pd.to_numeric(inact_long["Inactivity_Rate_Percent"], errors="coerce")
    inact_long["Year"] = inact_long["Year"].astype(int)
    inact_long = inact_long.dropna(subset=["Inactivity_Rate_Percent"]).reset_index(drop=True)

    if inact_long[["Area", "Year"]].duplicated().any():
        inact_long = (
            inact_long.groupby(["Area", "Year"], as_index=False)["Inactivity_Rate_Percent"].mean()
        )

    return inact_long


def build_hp_long(path, value_name):
    """Build long-format house price to earnings ratio data."""
    xls = pd.ExcelFile(path)
    df = pd.read_excel(xls, sheet_name="Median Earnings to Prices ratio")
    df = df.dropna(subset=[df.columns[0]])

    original_year_cols = [col for col in df.columns if isinstance(col, (int, float))]

    df = df.rename(columns={df.columns[0]: "Code", df.columns[2]: "Area"})

    df['Area'] = df['Area'].astype(str).str.replace('City OF LONDON', 'City of London').str.strip()
    df = df[df['Area'].isin(LONDON_BOROUGH_NAMES)].copy()

    hp_long = df.melt(
        id_vars=["Code", "Area"],
        value_vars=original_year_cols,
        var_name="Year",
        value_name=value_name
    )

    hp_long["Year"] = hp_long["Year"].astype(float).astype(int)
    hp_long[value_name] = pd.to_numeric(
        hp_long[value_name].astype(str).str.replace(r'[#,-,[\[x\]]', '', regex=True),
        errors="coerce"
    )

    hp_long = hp_long.dropna(subset=[value_name])

    if not hp_long[['Area', 'Year']].duplicated().any():
        return hp_long
    else:
        return hp_long.groupby(['Area', 'Year'], as_index=False)[value_name].mean()


# -------------------- File Paths --------------------
pop_path = os.path.join(drive_path, 'Population estimates.csv')
birth_path = os.path.join(drive_path, 'Live births in England and Wales.csv')
earn_xls_path = os.path.join(drive_path, "earnings-workplace-borough.xls")
inact_xls_path = os.path.join(drive_path, "economic-inactivity-by-gender-borough.xlsx")
hp_res_xls_path = os.path.join(drive_path, "ratio-house-price-earnings-residence-based.xlsx")
hp_work_xls_path = os.path.join(drive_path, "ratio-house-price-earnings-workplace-based.xlsx")

LAD_PREFIX = r'(lacu2023|ladu2023|ualad19|ualad14|ladu):'

# -------------------- 2. Load and Convert Data to Long Format --------------------

pop_df = clean_and_melt_csv(pop_path, skip_rows=6, measure_name='Population', lad_prefix_pattern=LAD_PREFIX)
birth_df = clean_and_melt_csv(birth_path, skip_rows=6, measure_name='Live_Births', lad_prefix_pattern=LAD_PREFIX)
earn_long = build_earn_long(earn_xls_path)
inact_long = build_inact_long(inact_xls_path)

print("✅ inact_long shape:", inact_long.shape)
print(inact_long.head())
print(inact_long.groupby("Area")["Year"].agg(["min", "max"]).head(10))

hp_res_long = build_hp_long(hp_res_xls_path, "HP_Residence_Ratio")
hp_work_long = build_hp_long(hp_work_xls_path, "HP_Workplace_Ratio")

# -------------------- 3. Merge All Data --------------------

final_merged_df = (
    pop_df
    .merge(birth_df, on=['Area', 'Year'], how='outer')
    .merge(earn_long, on=['Area', 'Year'], how='left')
    .merge(inact_long, on=['Area', 'Year'], how='left')
    .merge(hp_res_long, on=['Area', 'Year'], how='left')
    .merge(hp_work_long, on=['Area', 'Year'], how='left')
)

# Filter London boroughs and remove redundant Code columns
final_merged_df = final_merged_df[final_merged_df['Area'].isin(LONDON_BOROUGH_NAMES)].copy()
final_merged_df = final_merged_df.drop(columns=[col for col in final_merged_df.columns if "Code" in col and col != "postcode"], errors="ignore")

final_merged_df = final_merged_df.sort_values(['Area', 'Year']).reset_index(drop=True)

print("--- Final merged dataset (first 20 rows sorted by Area and Year) ---")
print(final_merged_df.head(20).to_markdown(index=False))
```
